```python
import numpy as np 
from sklearn.preprocessing import add_dummy_feature
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

m=100

x_gd=np.random.rand(m,1)
y=3*x_gd+np.random.randn(m,1)
x_b=add_dummy_feature(x_gd)
l_r=0.1
n_epochs=100
theta=np.random.randn(2,1)


 





    
        





    

```

```python
def batch_gd(x,y,l_r,n_epochs):
    
        x=np.array(x)
        y=np.array(y)
        theta=np.random.randn(2,1)
        costs=[]
        for epochs in range(n_epochs):
            gradients=2/m * x.T @(x@theta-y)
            theta=theta-l_r*gradients
            print(theta)
            y_pred=x_b@theta
            cost=mean_squared_error(y_pred,y)
            costs.append(cost)
            
        return costs,theta
```

```python
costs,last_theta=batch_gd(x_b,y,l_r,n_epochs)
plt.plot(costs)
plt.xlabel("Epochs")
plt.ylabel("Error(MSE)")
plt.title("Cost Curve of Batch GD")
plt.show()
```

**Observation**
so yes the gradient descent is doing its job as learning rate is 0.1

```python
costs,last_theta=batch_gd(x_b,y,0.01,n_epochs)
plt.plot(costs)
plt.xlabel("Epochs")
plt.ylabel("Error(MSE)")
plt.title("Cost Curve of Batch GD")
plt.show()
```

**observation**
here the learning rate is 0.01 and yes this thing did show its magic as the less rate means it wont reduce the cost in a very steep manner unlike 0.1 did before as it was a L shaped curve but now its a less steep curve.
